```python
from matplotlib import pyplot as plt
import numpy as np


#Code to plot the motion of an object that changes speed over time


# Setting values for time (in seconds)
t = np.linspace(0, 100, 100)
t = [0,1,2,3,4,5,6]

t = np.linspace(0, 10, 100)


t = [0, 0.1, 0.2 ]

t[3]
# Setting initial values for distance (all 0)
distance = np.zeros(100)

# Defining the speed of the object at different points in time
speed1 = 10
speed2 = 30

#defining the distance to speed relationship for the first 50 seconds
#add in a 'for loop' to calculate the distance values at each second for the first 50 seconds
for i in range(0,50):
    distance[i] = speed1*t[i]

#defining the distance to speed relationship for the last 50 seconds
#use a similar 'for loop'
for i in range(50,100):
    distance[i] = distance[49] + speed2*(t[i]-t[49])


# Create a plot to visualise the data you've specified
plt.plot(t, distance)

# Add labels and title to the plot
plt.xlabel("Time (Seconds)")
plt.ylabel("Distance (Metres)")
plt.title("Distance vs Time plot for object which changes speed")

# Show the plot
plt.show()

```
